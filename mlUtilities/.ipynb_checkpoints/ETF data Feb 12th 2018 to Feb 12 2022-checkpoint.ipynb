### ETF Data Exploratory Work

```python
# Libraries
import pandas as pd
import numpy as np
import datetime # imports the datetime library
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the data sets
spy = pd.read_csv("SPY.csv")
xlb = pd.read_csv("XLB.csv")
xlc = pd.read_csv("XLC.csv")
xle = pd.read_csv("XLE.csv")
xlf = pd.read_csv("XLF.csv")
xli = pd.read_csv("XLI.csv")
xlk = pd.read_csv("XLK.csv")
xlp = pd.read_csv("XLP.csv")
xlre = pd.read_csv("XLRE.csv")
xlu = pd.read_csv("XLU.csv")
xlv = pd.read_csv("XLV.csv")
xly = pd.read_csv("XLY.csv")
```

```python
# Add an additional column to each dataframe with ETF name.
spy["ETF_ticker"] = 'spy'
xlb["ETF_ticker"] = 'xlb'
xlc["ETF_ticker"] = 'xlc'
xle["ETF_ticker"] = 'xle'
xlf["ETF_ticker"] = 'xlf'
xli["ETF_ticker"] = 'xli'
xlk["ETF_ticker"] = 'xlk'
xlp["ETF_ticker"] = 'xlp'
xlre["ETF_ticker"] = 'xlre'
xlu["ETF_ticker"] = 'xlu'
xlv["ETF_ticker"] = 'xlv'
xly["ETF_ticker"] = 'xly'
```

```python
print(spy.shape)
print(xlb.shape)
print(xlc.shape)
print(xle.shape)
print(xlf.shape)
print(xli.shape)
print(xlk.shape)
print(xlp.shape)
print(xlre.shape)
print(xlu.shape)
print(xlv.shape)
print(xly.shape)
```

```python
# Create a list with al of the datasets
etf_dfs = [spy, xlb, xlc, xle, xlf, xli, xlk, xlp, xlre, xlu, xlv, xly]
```

#### Concatenate all of the dataframes

```python
etfs_combined = pd.concat(etf_dfs)
```

```python
etfs_combined.head()
```

#### Sort rows by date and ETF_ticker

```python
# Use the sort function to sort by date and ETF_ticker
etfs_sorted = etfs_combined.sort_values(by=['Date', 'ETF_ticker'])
etfs_sorted
```

#### Create dummy variables for the ticker variables

```python
pd.get_dummies(etfs_sorted, columns = ['ETF_ticker'])
```

#### Display ETFs as time series based on ticker name 

```python
sns.set(style="darkgrid")

sns.lineplot(data = pd.melt(etfs_sorted, ["Date","Close", "ETF_ticker"]), x = "Date", y = "Close", hue= "ETF_ticker")
```

#### Send to CSV

```python
etfs_combined.to_csv("combined_etfs.csv")
etfs_sorted.to_csv("sorted_etfs.csv")
```
